## 1. Stemming vs Lemmatization

### Objective
To process raw text and compare stemming and lemmatization techniques.

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import word_tokenize

# Input text
text = "I am running and studying better methods for learning"

# Tokenize
words = word_tokenize(text)

# Initialize tools
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

print("Original Words:", words)

print("\n--- Stemming ---")
for word in words:
    print(word, "→", stemmer.stem(word))

print("\n--- Lemmatization ---")
for word in words:
    print(word, "→", lemmatizer.lemmatize(word, pos='v'))
```

### N-grams and Next Word Prediction

```python
from nltk.util import ngrams

tokens = word_tokenize(text)

# Generate n-grams
unigrams = list(ngrams(tokens, 1))
bigrams = list(ngrams(tokens, 2))
trigrams = list(ngrams(tokens, 3))

print("Unigrams:\n", unigrams)
print("\nBigrams:\n", bigrams)
print("\nTrigrams:\n", trigrams)
```

```python
from collections import Counter

bigram_model = Counter(bigrams)

def predict_next(word):
    candidates = {pair[1]: count for pair, count in bigram_model.items() if pair[0] == word}
    return max(candidates, key=candidates.get) if candidates else "No prediction"

print("\nNext word after 'playing':", predict_next("playing"))
```

### Comparison

**Unigram Model:**
- Uses single words
- No context
- Least accurate

**Bigram Model:**
- Uses pairs of words
- Moderate context
- Better predictions

**Trigram Model:**
- Uses three-word sequences
- High context
- Most accurate but needs more data

### Conclusion

N-gram models improve prediction accuracy as the value of n increases. However, higher-order models require more data and computational power.

## 5. Parsing (CKY vs Earley)

### Objective
To parse sentences using context-free grammar and compare CKY and Earley parsing methods.

```python
from nltk import CFG
from nltk.parse import ChartParser

# Define grammar
grammar = CFG.fromstring("""
S -> NP VP
NP -> 'John'
VP -> V NP
V -> 'eats'
NP -> 'apple'
""")

# Create parser
parser = ChartParser(grammar)

sentence = "John eats apple".split()

print("Parse Tree:\n")
for tree in parser.parse(sentence):
    print(tree)
```

### Explanation

Parsing is used to analyze the grammatical structure of a sentence.

In this example:
- S → Sentence
- NP → Noun Phrase
- VP → Verb Phrase

The parser builds a tree representing sentence structure.

### Comparison

**CKY Parser:**
- Uses dynamic programming
- Works only with Chomsky Normal Form (CNF)
- Efficient for structured grammars

**Earley Parser:**
- More flexible
- Works with any CFG
- Slightly slower but more powerful

### Conclusion

CKY parsing is efficient but restricted to CNF grammars, while Earley parsing is more flexible and suitable for general-purpose parsing.

## 6. Dependency Parsing

### Objective
To extract subject–verb–object (SVO) relationships from a sentence and compare dependency parsing with constituency parsing.

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = "John eats an apple"

doc = nlp(text)

print("Token | Dependency | Head")
print("--------------------------------")

for token in doc:
    print(f"{token.text:10} | {token.dep_:10} | {token.head.text}")
```

```python
subject = ""
verb = ""
obj = ""

for token in doc:
    if token.dep_ == "nsubj":
        subject = token.text
    elif token.dep_ == "ROOT":
        verb = token.text
    elif token.dep_ == "dobj":
        obj = token.text

print("\nExtracted SVO:")
print("Subject:", subject)
print("Verb:", verb)
print("Object:", obj)
```

### Comparison

**Dependency Parsing:**
- Focuses on relationships between words
- Directly extracts SVO structure
- Easier for information extraction

**Constituency Parsing:**
- Breaks sentence into phrases
- Focuses on structure (NP, VP)
- More complex to interpret

### Conclusion

Dependency parsing is more useful for extracting relationships like subject–verb–object, while constituency parsing is better for understanding hierarchical sentence structure.

### Explanation

CFG generates sentences using fixed rules.

PCFG assigns probabilities to each rule, making sentence generation more realistic and data-driven.

### Comparison

**CFG (Context-Free Grammar):**
- Fixed rules
- No probabilities
- Deterministic output

**PCFG (Probabilistic CFG):**
- Rules have probabilities
- Produces more realistic language
- Handles ambiguity better

### Conclusion

PCFG is more powerful than CFG as it incorporates probabilities, making it suitable for real-world NLP tasks.

## 11. Pronoun Resolution

### Objective
To resolve pronouns in text and identify the noun they refer to.

```python
text = "John went to the store. He bought milk."

sentences = text.split(".")

# Simple heuristic: assume pronoun refers to last noun
noun = ""
pronoun_resolution = {}

for sentence in sentences:
    words = sentence.split()
    for word in words:
        if word.lower() in ["he", "she", "it", "they"]:
            pronoun_resolution[word] = noun
        else:
            noun = word

print("Pronoun Resolution:\n")
for pronoun, ref in pronoun_resolution.items():
    print(f"{pronoun} → {ref}")
```

### Explanation

Pronoun resolution identifies the noun that a pronoun refers to.

Example:
- "John went to the store. He bought milk."
- "He" refers to "John"

### Comparison

**Heuristic-Based Approach:**
- Simple rules
- Fast and easy
- Not very accurate

**Deep Learning-Based Approach:**
- Uses neural networks
- Highly accurate
- Requires large datasets

### Conclusion

Heuristic methods are simple but limited, while deep learning approaches provide better accuracy in resolving pronouns.

```python
# 1) Stemming vs Lemmatization: speed + simple 'dictionary' accuracy heuristic
def compare_stem_lemmatize(texts):
    results = []
    for text in texts:
        tokens = word_tokenize(text)
        # time stemming
        t0 = time.perf_counter()
        stems = [stemmer.stem(t) for t in tokens]
        t1 = time.perf_counter()
        # time lemmatization (verb pos as heuristic)
        t2 = time.perf_counter()
        lemmas = [lemmatizer.lemmatize(t, pos='v') for t in tokens]
        t3 = time.perf_counter()
        # simple accuracy proxy: how many outputs exist as WordNet lemmas (any synset)
        def wn_presence(words):
            return sum(1 for w in words if wn.synsets(w))
        stem_dict_hits = wn_presence(stems)
        lemma_dict_hits = wn_presence(lemmas)
        results.append({
            'text': text,
            'n_tokens': len(tokens),
            'stem_time_s': t1 - t0,
            'lemma_time_s': t3 - t2,
            'stem_dict_hits': stem_dict_hits,
            'lemma_dict_hits': lemma_dict_hits,
            'stems': stems,
            'lemmas': lemmas
        })
    return results

# Run on our SAMPLES and display a compact table-like summary
res = compare_stem_lemmatize(SAMPLES)
for r in res:
    print('TEXT:', r['text'])
    print(' tokens=', r['n_tokens'], ' stem_time=', round(r['stem_time_s'],6), 'lemma_time=', round(r['lemma_time_s'],6))
    print(' dictionary hits -> stems:', r['stem_dict_hits'], 'lemmas:', r['lemma_dict_hits'])
    print(' stems:', r['stems'])
    print(' lemmas:', r['lemmas'])
    print('---')
```

```python
# 2) NER: rule-based vs spaCy (ML) on a tiny annotated set
ANNOTATED = [
    {'text': 'Elon Musk founded Tesla in California in 2003.', 'ents': [('Elon Musk','PERSON'), ('Tesla','ORG'), ('California','GPE')]},
    {'text': 'Apple released the new iPhone in September.', 'ents': [('Apple','ORG')]},
    {'text': 'Barack Obama was born in Hawaii.', 'ents': [('Barack Obama','PERSON'), ('Hawaii','GPE')]},
]

def rule_based_ner(text):
    # Very simple rules: Capitalized sequences and known keywords
    ents = []
    # capitalized words sequences (naive)
    for match in re.finditer(r'\b([A-Z][a-z]+(?:\s+[A-Z][a-z]+)*)\b', text):
        span = match.group(1)
        # classify heuristically
        if any(k in span.lower() for k in ['inc','ltd','corp','company','group','university','hospital','college']):
            label='ORG'
        elif span.lower() in ['monday','tuesday','january','february']:
            label='DATE'
        else:
            # treat short person-like names as PERSON, others GPE/ORG depending on tokens
            label='PERSON' if len(span.split())<=2 else 'GPE'
        ents.append((span,label))
    return ents

def evaluate_ner():
    from collections import defaultdict
    stats = defaultdict(lambda: {'tp':0,'fp':0,'fn':0})
    for item in ANNOTATED:
        text=item['text']
        gold=set(item['ents'])
        rule=set(rule_based_ner(text))
        ml=set()
        try:
            doc=nlp(text)
            ml=set((ent.text, ent.label_) for ent in doc.ents)
        except Exception:
            ml=set()
        for pred,label in list(rule):
            if (pred,label) in gold: stats['rule']['tp']+=1
            else: stats['rule']['fp']+=1
        for g in gold:
            if g not in rule: stats['rule']['fn']+=1
        for pred,label in list(ml):
            if (pred,label) in gold: stats['ml']['tp']+=1
            else: stats['ml']['fp']+=1
        for g in gold:
            if g not in ml: stats['ml']['fn']+=1
    # compute precision/recall
    for k in ['rule','ml']:
        tp=stats[k]['tp']; fp=stats[k]['fp']; fn=stats[k]['fn']
        prec = tp/(tp+fp) if (tp+fp)>0 else 0
        rec = tp/(tp+fn) if (tp+fn)>0 else 0
        print(f'{k} NER -> TP={tp} FP={fp} FN={fn} Precision={prec:.2f} Recall={rec:.2f}')

evaluate_ner()
```

```python
# 4) Simple CKY (CYK) parser for CNF grammars and comparison with NLTK Chart/Earley parsers
def cyk_parse(words, productions, start='S'):
    # productions: dict lhs -> list of rhs tuples (each rhs is tuple of symbols)
    n = len(words)
    table = [[set() for _ in range(n)] for _ in range(n)]
    # initialize with preterminal rules A -> 'a'
    for i,w in enumerate(words):
        for A, rhss in productions.items():
            for rhs in rhss:
                if len(rhs)==1 and rhs[0]==w: table[i][i].add(A)
    # fill table
    for l in range(2, n+1):
        for i in range(n-l+1):
            j = i + l - 1
            for k in range(i, j):
                for A, rhss in productions.items():
                    for rhs in rhss:
                        if len(rhs)==2:
                            B, C = rhs
                            if B in table[i][k] and C in table[k+1][j]:
                                table[i][j].add(A)
    return (start in table[0][n-1]) if n>0 else False

# tiny CNF grammar for demo
productions = {
    'S': [('NP','VP')],
    'NP': [('John',), ('Mary',)],
    'VP': [('V','NP')],
    'V': [('eats',)],
}
words = 'John eats Mary'.split()
print('CYK accepts:', cyk_parse(words, productions))
# NLTK comparison (ChartParser used earlier); Earley parser may be available in nltk.parse
from nltk import CFG
from nltk.parse import ChartParser
grammar = CFG.fromstring("""S -> NP VP
NP -> 'John' | 'Mary'
VP -> V NP
V -> 'eats'""")
parser = ChartParser(grammar)
print('NLTK Chart parser parses:')
for tree in parser.parse(words):
    print(tree)
```

```python
# 5) Dependency parsing -> SVO extraction (wrapper around spaCy)
def extract_svo_spacy(text):
    try:
        doc = nlp(text)
    except Exception:
        return ()
    subj = verb = obj = ''
    for token in doc:
        if token.dep_=='nsubj': subj = token.text
        if token.dep_=='ROOT': verb = token.text
        if token.dep_=='dobj': obj = token.text
    return (subj, verb, obj)

print('SVO for "John eats an apple":', extract_svo_spacy('John eats an apple'))
```

```python
# 6) PCFG sampling helper (sample probabilistically)
import random
from nltk import PCFG

def sample_pcfg(grammar_text, n=5):
    grammar = PCFG.fromstring(grammar_text)
    # build production map with probabilities
    prods = {}
    for p in grammar.productions():
        lhs = str(p.lhs())
        prods.setdefault(lhs, []).append((p.rhs(), p.prob()))
    def expand(symbol):
        if symbol not in prods: return [symbol]
        rhss = prods[symbol]
        choices, probs = zip(*rhss)
        rhs = random.choices(choices, weights=probs, k=1)[0]
        out = []
        for sym in rhs:
            if isinstance(sym, str):
                out += expand(sym)
            else:
                out += expand(str(sym))
        return out
    samples=[]
    for _ in range(n):
        sent = expand(str(grammar.start()))
        samples.append(' '.join(sent))
    return samples

pcfg_text = """S -> NP VP [1.0]
NP -> 'John' [0.5] | 'Mary' [0.5]
VP -> V NP [1.0]
V -> 'likes' [0.6] | 'hates' [0.4]
"""
print('PCFG samples:', sample_pcfg(pcfg_text, n=5))
```

```python
# 7) WSD: Lesk (dictionary) vs small supervised classifier (toy dataset)
from nltk.wsd import lesk
from nltk.classify import NaiveBayesClassifier

TRAIN = [
    ('I deposited money in the bank', 'finance'),
    ('The river bank was full of fish', 'river'),
    ('She withdrew cash from the bank', 'finance'),
    ('He sat on the bank and watched the water', 'river'),
]

def features_for_wsd(context, word):
    toks = [t.lower() for t in word_tokenize(context) if t.isalpha()]
    feats = {}
    for t in toks:
        feats[f'has({t})'] = True
    return feats

train_feats = [(features_for_wsd(s, 'bank'), label) for s, label in TRAIN]
classifier = NaiveBayesClassifier.train(train_feats)

test_sent = 'I went to the bank to deposit money'
ls = lesk(word_tokenize(test_sent), 'bank')
print('Lesk sense (brief):', ls.definition() if ls else None)
pred = classifier.classify(features_for_wsd(test_sent, 'bank'))
print('Toy supervised prediction:', pred)
```

```python
# 10) Pronoun resolution: heuristic vs placeholder for deep model
def heuristic_pronoun_resolution(text):
    sentences = [s.strip() for s in re.split(r'(?<=[.!?])\s+', text) if s.strip()]
    last_noun = None
    res = {}
    for s in sentences:
        toks = word_tokenize(s)
        for t in toks:
            if t.istitle(): last_noun = t
            if t.lower() in ['he','she','they','it'] and last_noun:
                res[t]=last_noun
    return res

text = 'John went to the store. He bought milk.'
print('Heuristic pronoun resolution:', heuristic_pronoun_resolution(text))
print('(For deep-learning based coreference resolution, consider using a coref model such as NeuralCoref or a transformer-based pipeline.)')
```
